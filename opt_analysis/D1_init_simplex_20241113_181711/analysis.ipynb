# Results analysis for RB optimization on qubit D1

The optimization was run using NelderMead, the parameters on which optimization was performed are frequency and amplitude for the RX pulse.
In this case also an initial simplex was fixed for running the optimization.

I don't remember the platform configuration from which I started this optimization, in particular (even if in this case may not be relevant), the DRAG pulse was note previously tuned

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import json
```

```python
WIDTH = 0.5
FONT = 24

BLUE = "#788ae3"
RED = "#d1545e"
DARKBLUE = "#075cb3"
DARKRED = "#8c1822"
```

```python
loaded = np.load("optimization_history.npz")
iterations = loaded["iterations"]
parameters = loaded["parameters"]
objective_values = loaded["objective_values"]
objective_values_error = loaded["objective_value_errors"]
```

```python
# Plot each parameter as a function of the iteration step
params_name = ["amplitude [a.u.]", "frequency [Hz]"]

for i in range(parameters.shape[1]):
    plt.figure(figsize=(10, 6))  # Create a new figure for each parameter
    plt.plot(iterations, parameters[:, i], label=f"{params_name[i]}")
    plt.xlabel("# optimization step")
    plt.ylabel(f"{params_name[i]}")
    plt.title("Parameter Value")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Create a new figure for each parameter
plt.errorbar(iterations, 1 - objective_values, yerr=objective_values_error, fmt="o", markersize=4,
    color="black", linestyle="-", markerfacecolor=RED, markeredgewidth=0.7, linewidth=0.7, alpha=1,)
plt.xlabel("# optimization step")
plt.ylabel("Parameter Value")
plt.yscale("log")
plt.title(f"Fidelity")
plt.legend()
plt.grid()
plt.show()
```
